```python
import os
import openai
openai.api_key = "API KEY"
```

```python
#takes a input in english and returns the translation in french
def translate():
    inp = input("Enter your English text: ")
    q = f"English: {inp} \n French"
    response = openai.Completion.create(
        engine="davinci",
        prompt=q,
        temperature=0.5,
        max_tokens=100,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0,
        stop ='\n')
    french = response['choices'][0]['text'].split(':')[1]
    print("--------------------------------")
    print("--------------------------------")
    print(f"French Translation: {french} ")
```

```python
translate()
```
